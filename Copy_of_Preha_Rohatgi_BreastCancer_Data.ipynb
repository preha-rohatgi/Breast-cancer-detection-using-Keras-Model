```python
### Download data from google drive. You need not mess with this code.

import requests

def download_file_from_google_drive(id, destination):
    URL = "https://docs.google.com/uc?export=download"

    session = requests.Session()

    response = session.get(URL, params = { 'id' : id }, stream = True)
    token = get_confirm_token(response)

    if token:
        params = { 'id' : id, 'confirm' : token }
        response = session.get(URL, params = params, stream = True)

    save_response_content(response, destination)    

def get_confirm_token(response):
    for key, value in response.cookies.items():
        if key.startswith('download_warning'):
            return value

    return None

def save_response_content(response, destination):
    CHUNK_SIZE = 32768

    with open(destination, "wb") as f:
        for chunk in response.iter_content(CHUNK_SIZE):
            if chunk: # filter out keep-alive new chunks
                f.write(chunk)
                
if __name__ == "__main__":
    file_id = '1ih8PomVE7L3z_xReHAEsq3hk-0O1Uo12'
    destination = 'data.csv'
    download_file_from_google_drive(file_id, destination)
```

```python
def convert(y):

  ## Convert the output labels to numbers : M->0; B-> 1
  ## Store the output in Y_v
  Y_v = []

  #   YOUR CODE HERE

  for i in y:
    if(i=='B'):
      Y_v.append(1)
    else:
      Y_v.append(0)
  return(Y_v)  
```

```python
### One-hot encode Y_v
def oneHot(y, Ny):
    '''
    Input:
        y: an int in {0, 1}
        Ny: Number of classes, e.g., 2 here.
    Output:
        Y: a vector of Ny (=2) tuples
    '''
    # YOUR CODE HERE
    
    Y = to_categorical(y, Ny)
    
    return(Y)
    
```

```python
import keras
# YOUR CODE HERE
from keras.layers import Input, Dense
from keras.models import Model
print(X.shape[0])
input_layer = Input(shape = (X_train.shape[1:]))
hidden_layer_1 = Dense(64, activation = 'sigmoid')(input_layer)
hidden_layer_2 = Dense(512, activation = 'tanh')(hidden_layer_1)
output_layer = Dense(2, activation = 'softmax')(hidden_layer_2)

model = Model(inputs = [input_layer], outputs = [output_layer])

##compiling the model
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy',  metrics = ['accuracy'])
model.summary()
```
